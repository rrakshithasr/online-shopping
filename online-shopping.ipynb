```python
import pandas as pd

# -----------------------------
# Create Sample Dataset
# -----------------------------
data = {
    "Order_ID": [101,102,103,104,105,106,107,108,109,110],
    "Customer": ["Rahul","Priya","Aman","Neha","Arjun","Sneha","Kiran","Pooja","Rohan","Meena"],
    "Product": ["Laptop","Mouse","Keyboard","Monitor","Headphones","Webcam","Printer","SSD","RAM","Speaker"],
    "Quantity": [1,2,1,1,2,1,1,2,2,1],
    "Price": [55000,500,1500,12000,2500,3000,9000,4500,3500,2000],
    "Status": ["Delivered","Pending","Delivered","Delivered","Pending",
               "Delivered","Delivered","Pending","Delivered","Delivered"]
}

df = pd.DataFrame(data)

# Save dataset as CSV
df.to_csv("orders.csv", index=False)

print("="*60)
print("ONLINE SHOPPING ORDERS ANALYSIS")
print("="*60)

# -----------------------------
# Read CSV File
# -----------------------------
df = pd.read_csv("orders.csv")

print("\n1. Complete Dataset")
print(df)

print("\n2. First 5 Rows")
print(df.head())

print("\n3. Last 5 Rows")
print(df.tail())

print("\n4. Dataset Information")
df.info()

print("\n5. Summary Statistics")
print(df.describe())

print("\n6. Dataset Shape")
print(df.shape)

print("\n7. Column Names")
print(df.columns.tolist())

print("\n8. Data Types")
print(df.dtypes)

# -----------------------------
# Add Total Amount Column
# -----------------------------
df["Total_Amount"] = df["Quantity"] * df["Price"]

print("\n9. Dataset with Total Amount")
print(df)

print("\n10. Total Revenue")
print("₹", df["Total_Amount"].sum())

print("\n11. Average Order Value")
print("₹", df["Total_Amount"].mean())

print("\n12. Highest Order Amount")
print(df.loc[df["Total_Amount"].idxmax()])

print("\n13. Lowest Order Amount")
print(df.loc[df["Total_Amount"].idxmin()])

print("\n14. Delivered Orders")
print(df[df["Status"] == "Delivered"])

print("\n15. Pending Orders")
print(df[df["Status"] == "Pending"])

print("\n16. Orders Sorted by Total Amount (Highest First)")
print(df.sort_values(by="Total_Amount", ascending=False))

print("\n17. Order Status Count")
print(df["Status"].value_counts())

print("\n18. Product Count")
print(df["Product"].value_counts())

print("\n19. Customers Who Purchased More Than One Item")
print(df[df["Quantity"] > 1])

# -----------------------------
# Save Updated Dataset
# -----------------------------
df.to_csv("updated_orders.csv", index=False)

print("\nUpdated dataset saved as 'updated_orders.csv'")

print("\nProject Completed Successfully!")
```
